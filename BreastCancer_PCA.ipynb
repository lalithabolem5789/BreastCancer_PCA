# Exploratory Data Anaysis

```python
# creating heatmap

plt.figure(figsize = (20, 12))
corr = df.corr()

sns.heatmap(corr, annot=True,linewidths=.5, cmap="Purples")

plt.show()
```

There are many columns which are highly correlated which are in dark shaded colors.

It causes MULTICOLLINEARITY.

We can reduce the MULTICOLLINEARITY between the features.

    - remove one or more highly correlated features.
    - standardize by subtracting their mean from each observation.
    - combine highly correlated features into one.
    - calculate Variance Inflation Factor(VIF) for every independent variable. VIF greater than 1.5 indicates       multicollinearity. Remove features with high VIF 
    - Perform Dimensionality Reduction techniques
        - PCA:-Principle Component Analysis is used when we want to reduce the no.of variables in our data but we are not sure about which variables to drop. PCA is a statistical technique used to reduce the dimensionality of a dataset while preserving as much information as possible. It achieves this by transforming the original features into a new set of orthogonal (uncorrelated) features called principal components. These components are ordered by the amount of variance they explain in the data.
        - LDA:-Linear Discriminant Analysis is a supervised learning algorithm used to classify a data set into multiple classes. It is a linear transformation technique which seeks to find a set of linear combinations of variables that best separate classes of data. LDA is a classification technique that looks at the data set and finds a linear combination of variables that will best separate the data. It is often used to reduce the number of variables in a data set while still retaining the important information for classification.

## Applying Principle Component Analysis (PCA)

```python
# creating features and label 

X1 = df.drop('diagnosis', axis = 1)    # features
y1 = df['diagnosis']   # target
```

```python
#feature scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X1)
```

```python
X_scaled
```

```python
from sklearn.decomposition import PCA


pca = PCA()
pca.fit(X_scaled)

plt.figure(1, figsize=(8, 7))
plt.clf()  # keeping the visualization clean
plt.plot(pca.explained_variance_ratio_, linewidth=2)
# explained_variance_ratio_ is the proportion of variance explained by each of the principal components (PCs) extracted from the data
plt.xlabel('n_components')
plt.ylabel('explained_variance_ratio_')
```

```python
# performing PCA

pca = PCA(n_components = 5)

X_pca = pca.fit_transform(X_scaled)
```

```python
# calculating explained variance
explained_variance = pca.explained_variance_ratio_
explained_variance
```

Higher values indicate that the corresponding principal component retains more information (variance) from the original dataset.

# Train Test Split

```python
X = X_pca
y = y1
```

```python
# splitting data into training and test set into 70% & 30% ratio

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
# shape of the train and test sets

print("Shape of training set:", X_train.shape)
print("Shape of test set:", X_test.shape)
```

# MACHINE LEARNING MODELS

## 1. Logistic Regression

```python
# fitting data to model

from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
```

```python
# model predictions

y_pred = log_reg.predict(X_test)
# use the trained logistic regression model to predict the outcomes (y_pred) for new or unseen data (X_test)
```

```python
# accuracy score

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print(accuracy_score(y_train, log_reg.predict(X_train)))

log_reg_acc = accuracy_score(y_test, log_reg.predict(X_test))
print(log_reg_acc)
```

```python
# confusion matrix

print(confusion_matrix(y_test, y_pred))
```

```python
# classification report

print(classification_report(y_test, y_pred))
```
